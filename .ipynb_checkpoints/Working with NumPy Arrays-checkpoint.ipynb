```python
#a.  The .npy format is a file format for storing NumPy arrays4 on disk. 
#There is a file named upb.npy in the data folder. Load the file into a NumPy array and display its shape.
#b. The data that you loaded in the previous step is a 3D NumPy array representing
# an RGB image. The shape of the array is (height, width, channels). Display
# the image using the matplotlib library.

import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
image_array = np.load("upb.npy")

# Display the shape
print("Shape of the array:", image_array.shape)

# Display the image
plt.imshow(image_array)
plt.axis("off")  # hides axis numbers
plt.show()
```

```python
# c. What is your interpretation of the .size attribute of this array?
print("Size of the array:", image_array.size)
```

```python
# d. Display a figure with three subplots (side by side in a single row) showing the red,
# green, and blue channels of the image separately.

import matplotlib.pyplot as plt

# Extract channels
red = image_array[:, :, 0]
green = image_array[:, :, 1]
blue = image_array[:, :, 2]

# Create subplots
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(red, cmap="Reds")
plt.title("Red Channel")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(green, cmap="Greens")
plt.title("Green Channel")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(blue, cmap="Blues")
plt.title("Blue Channel")
plt.axis("off")

plt.show()
```

```python
# e. Convert the image to grayscale by taking the average of the three channels. Display
# the grayscale image.

# Convert to grayscale
grayscale = image_array.mean(axis=2)

# Display grayscale image
plt.imshow(grayscale, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# f.Convert the image to a binary image by setting all pixel values less than 128 to 0
# and all pixel values greater than or equal to 128 to 255. Display the binary image.

import numpy as np

# Convert to binary
binary = np.where(grayscale < 128, 0, 255)

# Display binary image
plt.imshow(binary, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

```python
# Suppose we want to crop the image to a specific region. Crop the image to
# the region between the coordinates (100, 100) and (550, 700). Display the
# cropped image.

# Crop the image
cropped = image_array[100:550, 100:700]

# Display cropped image
plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```
